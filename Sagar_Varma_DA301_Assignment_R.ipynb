```r
# Create a new column indicating whether loyalty points are above or below the threshold of 3200
tr$loyalty_group <- ifelse(tr$loyalty_points >= 3200, "3200 and Above", "Below 3200")

# Calculate the counts for each group
loyalty_counts <- table(tr$loyalty_group)

# Convert the table to a data frame for plotting
loyalty_counts_df <- as.data.frame(loyalty_counts)

# Rename columns for clarity
colnames(loyalty_counts_df) <- c("Loyalty_Group", "Count")

# Calculate proportions
loyalty_counts_df$Proportion <- loyalty_counts_df$Count / sum(loyalty_counts_df$Count)

# Format proportions as percentages
loyalty_counts_df$Proportion_Percent <- paste0(round(loyalty_counts_df$Proportion * 100, 1), "%")

# Create the bar plot
ggplot(loyalty_counts_df, aes(x = Loyalty_Group, y = Count, fill = Loyalty_Group)) +
  geom_bar(stat = "identity") +
  geom_text(aes(label = paste0(Count, " (", Proportion_Percent, ")")), vjust = -0.5) + # Add count labels above each bar
  labs(title = "Number of Customers by Loyalty Points",
       x = "Loyalty Point Group",
       y = "Number of Customers") +
  theme_minimal()
```

Key Observations and Insights:

 - **Loyalty Point Distribution:** -
The distribution of loyalty points is right-skewed, indicating that most customers have lower loyalty points, while a smaller group has significantly higher points. This suggests a potential segmentation of customers based on loyalty levels.

 - **Outliers:** - There are outliers in the loyalty points data, particularly above 3200 points. These outliers represent customers with exceptionally high loyalty and should be investigated further to understand their behavior and characteristics.

 - **Spending Score and Loyalty Points:** - There is a positive correlation between spending score and loyalty points. Customers with higher spending scores tend to accumulate more loyalty points. This highlights the importance of encouraging spending to increase loyalty.

 - **Remuneration and Loyalty Points:** - There is a moderate positive correlation between remuneration (income) and loyalty points. While higher-income customers tend to have more loyalty points, the relationship is not as strong as with spending score.

 - **Age and Loyalty Points:**  - There is a weak positive correlation between age and loyalty points. Older customers tend to have slightly more loyalty points, but age is not a primary driver of loyalty point accumulation.

 - **Education and Loyalty Points:** - There is some variation in loyalty points across different education levels. Customers with higher education levels tend to have slightly higher loyalty points, but the difference is not substantial.

```r
# Create a Multiple Linear regression model named mlr using tr data and age, remuneration and spend_score as predictors for loyalty points

mlr <- lm(loyalty_points ~ age + remuneration + spend_score, data = tr)
summary(mlr)
```

```r
qqnorm(residuals(mlr))
qqline(residuals(mlr), col='blue')
```

```r
# Predict loyalty points using the mlr model
predictions <- predict(mlr, newdata = tr)

# Evaluate the model
# Calculate RMSE
rmse <- sqrt(mean((tr$loyalty_points - predictions)^2))
cat("RMSE:", rmse, "\n")

# Calculate R-squared
r_squared <- summary(mlr)$r.squared
cat("R-squared:", r_squared, "\n")
```

```r
# Model visualization (example: predicted vs actual)
plot(tr$loyalty_points, predict(mlr),
     xlab = "Actual Loyalty Points",
     ylab = "Predicted Loyalty Points",
     main = "Actual vs. Predicted Loyalty Points")
abline(0, 1, col = "red") # Add a 45-degree reference line
```

```r
# Perform k-means clustering for k = 4, 5, and 6
set.seed(123) # for reproducibility
km_result4 <- kmeans(tr[, c("spend_score", "remuneration")], centers = 4, nstart = 25)
km_result5 <- kmeans(tr[, c("spend_score", "remuneration")], centers = 5, nstart = 25)
km_result6 <- kmeans(tr[, c("spend_score", "remuneration")], centers = 6, nstart = 25)

# Compare the results (example: using total within-cluster sum of squares - WSS)
wss4 <- km_result4$tot.withinss
wss5 <- km_result5$tot.withinss
wss6 <- km_result6$tot.withinss

cat("Total within-cluster sum of squares (WSS):\n")
cat("k=4:", wss4, "\n")
cat("k=5:", wss5, "\n")
cat("k=6:", wss6, "\n")

#Further analysis using other metrics such as Silhouette width
sil_width4 <- silhouette(km_result4$cluster, dist(tr[, c("spend_score", "remuneration")]))
sil_width5 <- silhouette(km_result5$cluster, dist(tr[, c("spend_score", "remuneration")]))
sil_width6 <- silhouette(km_result6$cluster, dist(tr[, c("spend_score", "remuneration")]))

#Average Silhouette Width
mean(sil_width4[,3])
mean(sil_width5[,3])
mean(sil_width6[,3])
```

**Both elbow method and the silhouette point in the direction of k=5 as the most optimial value for clustering**

**Key Findings:**

They are predominantly young with moderate loyalty, indicating responsiveness to targeted promotions and engaging loyalty programs tailored to their preferences.

4. **Cluster 4 (High-Value Loyalists):** Comprising 18% of customers, this segment exhibits the highest spending scores, remuneration, and loyalty points. They are predominantly young, highlighting a crucial need for retention strategies focused on personalised experiences, premium rewards, and early access to new products.

5. **Cluster 5 (Average Consumers):** This largest segment (39% of customers) displays moderate spending scores, remuneration, and loyalty points. They are primarily middle-aged, suggesting an opportunity for targeted campaigns and personalised recommendations to enhance engagement and drive incremental spending.
